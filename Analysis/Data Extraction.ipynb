```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Data/Phase 0(T-2) studs..csv")
```

```python
df
```

```python
df1=df.copy()
```

```python
columns=['time','email','name','age','gender','physical','psychological','other_traits','student_pref','specialization','teacher_char','talents','other_talents','teacher_pref','life_love']
df1.columns=columns
```

```python
df1.drop(columns=['email','name','other_talents','life_love'],inplace=True)
```

```python
df1.info()
```

```python
df1['time']=pd.to_datetime(df1['time'])
```

```python
df1['physical']=df1['physical'].str.split(';')
df1['psychological']=df1['psychological'].str.split(';')
df1['other_traits']=df1['other_traits'].str.split(';')
```

```python
df1=df1.reset_index().rename(columns={'index':'ID'})
```

```python
def extract_trait(traits, column):
    for trait in traits[column]:
        df2.append([traits['ID'],trait])
```

```python
df2=[]
df1[['ID','physical']].apply(extract_trait, axis=1, args=('physical',))
df1[['ID','psychological']].apply(extract_trait, axis=1, args=('psychological',))
df1[['ID','other_traits']].apply(extract_trait, axis=1, args=('other_traits',))
user_traits=pd.DataFrame(df2,columns=['UserId','Trait'])
```

```python
user_traits
```

```python
df1['teacher_char']=df1['teacher_char'].str.split(';')
df1['talents']=df1['talents'].str.split(';')
```

```python
df2=[]
df1[~df1['teacher_char'].isna()][['ID','teacher_char']].apply(extract_trait, axis=1, args=('teacher_char',))
teachers_characters=pd.DataFrame(df2,columns=['UserId','TeacherChar'])
```

```python
teachers_characters
```

```python
df2=[]
df1[~df1['talents'].isna()][['ID','talents']].apply(extract_trait, axis=1, args=('talents',))
talents=pd.DataFrame(df2,columns=['UserId','Talents'])
```

```python
talents
```

```python
df1.drop(columns=['physical','psychological','other_traits','teacher_char','talents'],inplace=True)
```

```python
df1.head()
```

```python
user_traits.to_csv('../Data/UserTraits.csv',index=False)
teachers_characters.to_csv('../Data/TeacherChar.csv',index=False)
talents.to_csv('../Data/UserTalents.csv',index=False)
df1.to_csv('../Data/Main.csv',index=False)
```
